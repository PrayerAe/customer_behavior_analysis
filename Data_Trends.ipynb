```python
# Cek missing values

df.isnull().sum()
```

```python
df['review_rating'] = pd.to_numeric(df['review_rating'], errors='coerce')
```

```python
# Replace missing values dengan median karena kalau dengan mean akan sangat jauh angkanya untuk menghindari outliers
# Kita akan mengisi missing valuesnya mengambil dari per kategory untuk menghindari kategory yang berbeda diambil nilai mediannya yang akan menyebabkan bias pada data
# Jadi dari datasetnya jika ada missing values dari clothing item dia akan mengambil nilai median dari kategory clothing tidak akan mengambil dari kategory lainnya

df['Review Rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df['review_rating'].isna().sum()
df['review_rating'].dtype
```

```python
df.isnull().sum()
```

```python
# Mengubah nama kolom menjadi seragam semuanya (tidak ada spasi (digantikan underscore), tidak menggunakan huruf kapital.

# Mengubah semua kolom menjadi tidak ada huruf kapital
df.columns = df.columns.str.lower()

# Mengubah spasi menjadi underscore
df.columns = df.columns.str.replace(' ', '_')

# Mengubah nama kolom yang terlalu panjang
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

# FEATURE ENGINERING

```python
# Create a column age_group

labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
# qcut function adalah membagi data menjadiempat equal sized group berdasarkan dari data distribution dan memberikan label yang telah kita definisikan
```

```python
df[['age', 'age_group']].head(10)
```

```python
df['frequency_of_purchases'].head(10)
```

```python
# Membuat kolom purchase_frequency_days untuk mencari tahu seberapa sering seorang customer berbelanja
# Mengubah dari teks data harian pembelian (monthly, weeks, dst) menjadi jumblah angka bukan lagi text


frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# Menampilkan kolom untuk perbandingan

df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```
